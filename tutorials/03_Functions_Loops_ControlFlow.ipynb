## 1. For Loops - Repeating Operations

```python
# Simple loop over a list
temperatures = [15, 18, 22, 25, 20]

for temp in temperatures:
    print(f"Temperature: {temp}°C")

print("\n" + "="*40)

# Loop with range
print("Days 1 to 5:")
for day in range(1, 6):
    print(f"Day {day}")
```

## 2. Accumulating Values in Loops

A common pattern: calculate something step by step:

```python
# Calculate total rainfall
daily_precip = [5, 10, 2, 15, 8]
total_precip = 0

for precip in daily_precip:
    total_precip = total_precip + precip
    print(f"Added {precip}mm, total is now {total_precip}mm")

print(f"\nFinal total: {total_precip}mm")

# Same thing with multiplication
print("\n" + "="*40)
print("Calculating product:")
product = 1
numbers = [2, 3, 4]

for num in numbers:
    product = product * num
    print(f"Multiplied by {num}, product is now {product}")

print(f"Final product: {product}")
```

## 3. Loops with Conditions

Often you only want to do something in certain situations:

```python
# Count days with high rainfall
daily_precip = [5, 15, 2, 20, 8, 18, 3]
high_precip_days = 0

for precip in daily_precip:
    if precip > 10:
        print(f"High rainfall: {precip}mm")
        high_precip_days = high_precip_days + 1
    else:
        print(f"Normal rainfall: {precip}mm")

print(f"\nTotal days with high rainfall: {high_precip_days}")
```

## 4. While Loops - Repeat Until a Condition is Met

While loops are useful when you don't know how many iterations you need:

```python
# Example: Rain accumulation until we reach 50mm
import numpy as np

rainfall_data = [5, 8, 10, 12, 15, 6, 3]
total = 0
day = 0

while total < 50 and day < len(rainfall_data):
    total = total + rainfall_data[day]
    day = day + 1
    print(f"Day {day}: Total rainfall = {total}mm")

print(f"\nReached 50mm after {day} days")
```

## 6. Simulating a Simple Reservoir Model\n\nCombine loops and functions to simulate how a system changes over time:\n\nIn this example, we build a simple model of a water reservoir (dam) where water flows in from rainfall and out through release valves and evaporation.

```python
# Simple reservoir model
# Water flows in and out of a reservoir

def inflow_from_rainfall(rainfall_mm, catchment_area_km2):\n    \"\"\"\n    Calculate water inflow to reservoir from rainfall.\n    \n    Args:\n        rainfall_mm: Daily rainfall (mm = liters per square meter)\n        catchment_area_km2: Catchment area (km²)\n    \n    Returns:\n        inflow: Volume in million cubic meters\n    \"\"\"\n    # 1mm rainfall over 1km² = 1 million cubic meters\n    inflow = rainfall_mm * catchment_area_km2\n    return inflow\n\ndef outflow_from_release(release_fraction, current_volume):\n    \"\"\"\n    Calculate water released from reservoir.\n    \n    Args:\n        release_fraction: Fraction of water to release (0 to 1)\n        current_volume: Current reservoir volume (million m³)\n    \n    Returns:\n        outflow: Volume released (million m³)\n    \"\"\"\n    outflow = current_volume * release_fraction\n    return outflow\n\ndef evaporation_from_surface(current_volume, max_capacity, daily_potential_et):\n    \"\"\"\n    Calculate evaporation (water lost to atmosphere).\n    \n    Args:\n        current_volume: Current reservoir volume (million m³)\n        max_capacity: Maximum capacity (million m³)\n        daily_potential_et: Potential evapotranspiration (mm per day)\n    \n    Returns:\n        evap_loss: Volume lost to evaporation (million m³)\n    \"\"\"\n    # Larger reservoirs lose more water through evaporation\n    surface_area = (current_volume / max_capacity) ** 0.5  # Simplified\n    evap_loss = daily_potential_et * surface_area\n    return evap_loss\n\n# Simulation parameters\ndays = 30\ncatchment_area = 100  # km²\nmax_capacity = 50     # million m³\nrelease_fraction = 0.02  # Release 2% per day for power generation + supply\n\n# Rainfall data (mm/day)\nrainfall = [2, 5, 1, 8, 3, 0, 2, 10, 5, 2,\n            1, 3, 2, 4, 6, 1, 0, 2, 3, 5,\n            2, 1, 4, 3, 2, 1, 5, 2, 3, 1]\n\n# Start with 50% capacity\nvolume = max_capacity * 0.5\n\n# Storage for results\nvolumes = [volume]\ninflows = []\noutflows = []\nevaps = []\n\n# Run simulation\nprint(f\"Day | Volume | Inflow | Outflow | Evap | New Volume\")\nprint(\"-\" * 60)\n\nfor day in range(days):\n    # Calculate fluxes\n    inflow = inflow_from_rainfall(rainfall[day], catchment_area)\n    outflow = outflow_from_release(release_fraction, volume)\n    evap = evaporation_from_surface(volume, max_capacity, 2)  # 2mm/day potential ET\n    \n    # Update volume\n    new_volume = volume + inflow - outflow - evap\n    new_volume = max(0, min(new_volume, max_capacity))  # Keep within bounds\n    \n    # Store results\n    inflows.append(inflow)\n    outflows.append(outflow)\n    evaps.append(evap)\n    volumes.append(new_volume)\n    \n    # Print selected days\n    if day < 5 or day % 10 == 0:\n        print(f\"{day+1:3d} | {volume:6.1f} | {inflow:6.1f} | {outflow:7.2f} | {evap:4.2f} | {new_volume:6.1f}\")\n    \n    volume = new_volume\n\nprint(f\"\\nSimulation complete!\")\nprint(f\"Starting volume: {volumes[0]:.1f} million m³\")\nprint(f\"Ending volume: {volumes[-1]:.1f} million m³\")\nprint(f\"Total inflow: {sum(inflows):.1f} million m³\")\nprint(f\"Total outflow: {sum(outflows):.1f} million m³\")\nprint(f\"Total evaporation: {sum(evaps):.1f} million m³\")
```

## 6. Simulating Model Steps with Loops

Now let's combine loops and functions to simulate a simple water balance model:

```python
import numpy as np\n\n# Simple simulation with NumPy arrays\nnum_days = 100\n\n# Initialize arrays\ntemperature = np.zeros(num_days)\nair_temperature = np.zeros(num_days)\n\n# Set initial condition\ntemperature[0] = 5  # Cold starting temperature\n\n# Simulate seasonal air temperature (simple sine curve)\nfor i in range(num_days):\n    # Air temperature varies from 5°C to 25°C over the 100 days\n    air_temperature[i] = 15 + 10 * np.sin(i / 100 * 2 * np.pi)\n\n# Simulate water temperature\ncooling_rate = 0.1\n\nfor i in range(num_days - 1):\n    temp_diff = air_temperature[i] - temperature[i]\n    temperature[i + 1] = temperature[i] + temp_diff * cooling_rate\n\n# Summarize results\nprint(f\"100-day water temperature simulation:\")\nprint(f\"Starting temperature: {temperature[0]:.1f}°C\")\nprint(f\"Ending temperature: {temperature[-1]:.1f}°C\")\nprint(f\"Mean water temperature: {temperature.mean():.1f}°C\")\nprint(f\"Max water temperature: {temperature.max():.1f}°C\")\nprint(f\"Min water temperature: {temperature.min():.1f}°C\")
```
