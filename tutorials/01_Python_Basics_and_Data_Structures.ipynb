## 1. Variables and Data Types

In Python, a variable stores a value. Let's start simple:

```python
# Integers (whole numbers)
temperature = 25
rainfall = 100

# Floats (decimal numbers)
soil_moisture = 45.3
latitude = 50.5

# Strings (text)
location = "Germany"
season = "Spring"

# Booleans (True or False)
is_raining = True

print(f"Temperature: {temperature}°C")
print(f"Soil Moisture: {soil_moisture}%")
print(f"Location: {location}")
print(f"Is it raining? {is_raining}")
```

## 2. Basic Arithmetic

```python
# Simple calculations
precip_jan = 50  # mm
precip_feb = 45  # mm

total_precip = precip_jan + precip_feb
avg_precip = total_precip / 2
precip_per_hour = precip_jan / 24
precip_squaread = precip_jan ** 2

print(f"Total: {total_precip} mm")
print(f"Average: {avg_precip} mm")
print(f"Per hour: {precip_per_hour:.2f} mm/h")
print(f"Squared: {precip_squaread} mm²")
```

## 3. Lists - Storing Multiple Values

```python
# Create a list
temperatures = [15, 16, 18, 22, 25, 28, 29]
dates = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul"]

# Access elements (indexing starts at 0)
first_temp = temperatures[0]
last_temp = temperatures[-1]

print(f"First temperature: {first_temp}°C")
print(f"Last temperature: {last_temp}°C")

# Get the length
num_months = len(temperatures)
print(f"Number of months: {num_months}")

# Add elements
temperatures.append(27)
print(f"After adding: {temperatures}")
```

## 4. Dictionaries - Key-Value Pairs

```python
# A location with multiple properties
location_data = {
    'name': 'Germany',
    'latitude': 50.5,
    'longitude': 10.5,
    'elevation': 350,
    'annual_precip': 650
}

# Access values
print(f"Location: {location_data['name']}")
print(f"Latitude: {location_data['latitude']}")
print(f"Annual Precipitation: {location_data['annual_precip']} mm")

# Add a new key-value pair
location_data['climate'] = 'temperate'
print(f"Climate: {location_data['climate']}")
```

## 5. Conditionals (if/else)

Making decisions in your code:

```python
soil_moisture = 45

if soil_moisture > 70:
    print("Soil is very wet")
elif soil_moisture > 40:
    print("Soil moisture is good")
else:
    print("Soil is dry")

# Comparison operators
rainfall = 100

if rainfall > 80:
    status = "High rainfall"
else:
    status = "Normal rainfall"

print(status)
```

## 6. Writing Your First Function

Functions are reusable blocks of code:

```python
# Simple function to calculate water balance
def water_balance(precipitation, evaporation, runoff):
    """
    Calculate soil water balance.
    
    Args:
        precipitation: rainfall in mm
        evaporation: evaporation in mm
        runoff: water running off in mm
    
    Returns:
        net_change: change in soil moisture
    """
    net_change = precipitation - evaporation - runoff
    return net_change

# Use the function
precip = 50
evap = 20
runoff = 10

result = water_balance(precip, evap, runoff)
print(f"Water balance change: {result} mm")

# Call it again with different values
result2 = water_balance(80, 25, 15)
print(f"Water balance change: {result2} mm")
```
