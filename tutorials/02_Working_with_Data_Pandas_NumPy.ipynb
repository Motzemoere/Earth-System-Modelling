## 1. Introduction to pandas

Pandas is the main library for working with data in Python. Its core structure is the DataFrame:

```python
import pandas as pd
import numpy as np

# Create a simple DataFrame from a dictionary
weather_data = {
    'date': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'temperature': [10, 12, 8],
    'precipitation': [0.5, 2.3, 0.0],
    'location': ['Germany', 'Germany', 'Germany']
}

df = pd.DataFrame(weather_data)
print(df)
print(f"\nDataFrame shape: {df.shape}")  # (rows, columns)
```

## 2. Loading Data from CSV Files

Your project has CSV data files. Here's how to load them:

```python
# Load the Germany dataset
df_germany = pd.read_csv('data/Data_swbm_Germany.csv')

# Look at the first few rows
print("First 5 rows:")
print(df_germany.head())

# Get basic info
print("\nDataFrame info:")
print(df_germany.info())
```

## 3. Exploring Your Data

```python
# Get statistics
print(df_germany.describe())

# Get column names
print(f"\nColumns: {df_germany.columns.tolist()}")

# Get data types
print(f"\nData types:\n{df_germany.dtypes}")
```

## 4. Accessing Data

There are several ways to access data in a DataFrame:

```python
# Access a single column
precip = df_germany['tp_[mm]']
print(f"Precipitation column (first 5 values):\n{precip.head()}")

# Basic statistics on a column
print(f"\nTotal precipitation: {precip.sum():.1f} mm")
print(f"Mean precipitation: {precip.mean():.2f} mm")
print(f"Max precipitation: {precip.max():.2f} mm")
print(f"Min precipitation: {precip.min():.2f} mm")

# Access by row index
first_row = df_germany.iloc[0]
print(f"\nFirst row:\n{first_row}")
```

## 5. Filtering Data

```python
# Find days with high precipitation
high_precip = df_germany[df_germany['tp_[mm]'] > 10]
print(f"Days with precipitation > 10mm: {len(high_precip)} days")
print(high_precip[['time', 'tp_[mm]']].head())

# Find days with low soil moisture
low_sm = df_germany[df_germany['sm_[m3/m3]'] < 0.2]
print(f"\nDays with soil moisture < 0.2: {len(low_sm)} days")
```

## 6. Introduction to NumPy Arrays

NumPy is for fast numerical computations. It works with arrays:

```python
# Create arrays
temps = np.array([15, 16, 18, 22, 25, 28, 29])
precips = np.array([10, 15, 5, 20, 30, 25, 18])

print(f"Temperatures: {temps}")
print(f"Precipitation: {precips}")

# Element-wise operations
rainfall_to_precip_ratio = precips / 2
print(f"\nRatio (precip/2): {rainfall_to_precip_ratio}")

# Mathematical operations
evaporation = np.sin(temps / 10) * 5  # Simplified example
print(f"\nEvaporation (example): {evaporation.round(2)}")

# Common operations
print(f"\nSum of precipitation: {precips.sum()}")
print(f"Mean temperature: {temps.mean():.1f}")
print(f"Max temperature: {temps.max()}")
```

## 7. Converting Between Pandas and NumPy

You often need to convert between DataFrames and arrays:

```python
# Create a small DataFrame
small_df = pd.DataFrame({
    'precip': [10, 15, 20],
    'temp': [15, 18, 22]
})

# Convert column to NumPy array
precip_array = np.asarray(small_df['precip'])
print(f"Type: {type(precip_array)}")
print(f"Array: {precip_array}")

# Now perform fast NumPy operations
result = precip_array * 2
print(f"\nAfter multiplying by 2: {result}")
```
